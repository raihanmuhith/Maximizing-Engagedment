```python
import pandas as pd

df1 = pd.read_csv('mock_experimentation_log.csv')
df1 = df1.set_index('unit_id')
df1
```

```python
df2 = pd.read_csv('mock_homepage_carousel_log.csv')
df2 = df2.set_index('unit_id')
df2
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
alt_a = df3[df3['treatment'] == '{"variant":"alt_a"}']
alt_b = df3[df3['treatment'] == '{"variant":"alt_b"}']
control = df3[df3['treatment'] == '{"variant":"control"}']
print('total interaksi yang menggunakan algoritma a = ',alt_a[alt_a['event_name'] == 'CONTENTS ENGAGED'].shape[0])
print('total interaksi yang menggunakan algoritma b = ',alt_b[alt_b['event_name'] == 'CONTENTS ENGAGED'].shape[0])
print('total interaksi yang menggunakan algoritma control = ',control[control['event_name'] == 'CONTENTS ENGAGED'].shape[0])
```
